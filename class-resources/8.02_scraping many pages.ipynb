## Put everything together

```python
from random import uniform
pages = []
url_lst = []

for i in iterations:
    start_at = str(i)
    url = f"https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start={start_at}&ref_=adv_nxt"
    url_lst.append(url)
    
    # start_at needs to be a string if we used this:
    # url = "https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start=" + start_at + "&ref_=adv_nxt"
    
    # download the url
    response = requests.get(url)
    
    # print the status code
    print("Status code: " + str(response.status_code))
    
    pages.append(response)
    
    # respectful nap
    wait_time = uniform(0.3, 2)
    sleep(wait_time)
    
url_lst
```

```python
titles = []
synopsis = []
pages_parsed = []

# looping through all pages
for i in range(len(pages)):
    # a soup for ever search result page
    pages_parsed.append(BeautifulSoup(pages[i].content, "html.parser"))
    
    # movies
    movies_html = pages_parsed[i].select("div.lister-item-content")
    
    # loop through the html content of the movies of the current page (i)
    for j in range(len(movies_html)):
        
        # extract and append titles to title list
        titles.append(movies_html[j].select("h3 > a")[0].get_text())
        
        # extract and append synopsis to synopsis list
        synopsis.append(movies_html[j].select("p:nth-child(4)")[0].get_text())
        
```

```python
len(titles), len(synopsis)
```

```python
import pandas as pd
movies = pd.DataFrame({'titles': titles, 'synopsis': synopsis})
```

```python
movies['synopsis'] = movies['synopsis'].apply(lambda x: x.lstrip('\n').strip())
```
